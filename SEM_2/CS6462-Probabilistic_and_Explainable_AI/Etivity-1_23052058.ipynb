## Computing Probability

Use the ```fractions``` library, which provides support for rational number arithmetic. Use the ```compute_probability``` function to calculate probability. 

For more information on ```fractions```, please consult the reference page here: https://docs.python.org/3/library/fractions.html#fractions.Fraction

````python
from fractions import Fraction
import random

cases = len # the number of cases is the length, or size, of a set
favorable = set.intersection # outcomes that are in the event and in the sample space

def compute_probability(event, space): ### The probability of an event, given a sample space.

    result = Fraction(cases(favorable(event, space)), cases(space))
    probability = result.numerator / result.denominator
    
    return probability
````

# Step 3: Implement Events and Compute Probabilites.  (60%)

#### Utility function that will run the trials and return the combined probability

````python
def runTrials(eventSpace, sampleSpace, numberOfTrials):
    if numberOfTrials <= 0:
        return 0
        
    probability = 1
    bag = sampleSpace.copy() # Creating a copy of the diamonds set
    for trial in range(numberOfTrials):
        print(f'Diamonds in bag: {len(bag)}')
        print(f'Diamonds in event space: {len(eventSpace)}')
        probability *= compute_probability(eventSpace, bag)
        print(f'Probability of {trial+1} diamond(s) being drawn: {probability:.20f}')
        
        # remove random diamond from the event and space sets
        item_to_remove = random.choice(list(eventSpace))
        print(f'Removing diamond with properties {item_to_remove} from the bag.')
        eventSpace.remove(item_to_remove)
        bag.remove(item_to_remove)

    return probability
````

## Step 3.1: Compute the Probability of Drawing Five Diamonds with "Ideal" Cut Consecutively?   (15%)

<span style="color:red">TODO</span>

* implement the event "diamonds with ideal cut"
* compute the probability of drawing five diamonds with "ideal" cut consecutively and without replacement 

````python
ev_diamonds_with_ideal_cut = set()

#put your code here
for index, diamond in df.iterrows():
    if diamond['cut'] == 'Ideal': # Filtering for diamonds with ideal cut
        ev_diamonds_with_ideal_cut.add((index, diamond['cut'], diamond['color'], diamond['clarity']))

print(f'Sample size of event with diamonds of ideal cut being chosen: {len(ev_diamonds_with_ideal_cut)}')
# print(ev_diamonds_with_ideal_cut)

prob_ev_diamonds_with_ideal_cut = runTrials(ev_diamonds_with_ideal_cut, diamonds, 5)
print(f'Probability of drawing five diamonds with "ideal" cut consecutively and without replacement: {prob_ev_diamonds_with_ideal_cut}')
````
