## Visualising data

```python
import matplotlib.pyplot as plt
```

```python
plot_data = df.copy()
print(f'Before:\n{plot_data.head()}')
print('-' * 40)
plot_data = plot_data.set_index("Student")
print(f'After:\n{plot_data.head()}')
```

```python
plot_data.plot()
plt.show()
```

```python
df.iloc[:,0]
```

```python
plot_data["Physics"].plot(legend=True, 
                          title='Student Grades: Physics', 
                          ylabel='Grade')
plt.show()
```

```python
plt.plot(df.iloc[:,0], df.iloc[:,1], label='Physics')
# plt.ylabel('Grade')
plt.title('Student Grades: Physics')
plt.legend()
plt.show()
```

```python
plot_data.plot.area(figsize=(12, 4),
                    stacked=False,
                    ylabel='Grade', 
                    title='Class Grades: Area Plot (unstacked)')
plt.show()
```

```python
plot_data.plot.area(figsize=(12, 4),
                    subplots=True,
                    ylabel='Grade', 
                    title='Class Grades: Area Plot (subplots)')
plt.show()
```

```python
plot_data.plot.box(ylabel='Grade',
                   title='Class Grades: Box Plot')
plt.show()
```

```python
plot_data.plot.bar(ylabel='Grade', 
                   title='Class Grades: Bar Graph')
plt.show()
```

```python
plot_data.plot.barh(ylabel='Grade', 
                   title='Class Grades: Bar Graph - Horizontal')
plt.show()
```
