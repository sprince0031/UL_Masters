## Task 2.1

Downloading the shakespeare corpus for all our processing

```python
#TASK 2.1
!wget http://www.gutenberg.org/files/100/100-0.txt
filePath = "./100-0.txt"
shakespeareFile = open(filePath, "r")
shakespeareContent = shakespeareFile.read()
shakespeareContent = shakespeareContent.splitlines()
print(shakespeareContent[0:100])
```

## Task 2.2

Tokenising using Keras.

```python
import tensorflow as tf

# Initialising the tokeniser pre-processor
tok = tf.keras.preprocessing.text.Tokenizer(num_words=None, lower=True, split=" ")
tok.fit_on_texts(shakespeareContent)
print(tok.word_counts)
```

```python
# token count
print(f'a. Total token count in the corpus: {sum(tok.word_counts.values())}')
print(f'b. Total token types in the corpus: {len(tok.word_counts)}')
# Unnecessarily complicated this. Could have just taken max() value.
# typeListByMostFrequency = sorted(list(tok.word_counts.items()), key=lambda item: item[1], reverse=True)
print(f'c. Most frequent token type is "{max(tok.word_counts)}" in the corpus which has a frequency of {max(tok.word_counts.values())}.')
```

## Task 2.3

Stemming and lemmatizing

```python
from nltk.stem import PorterStemmer
import nltk
nltk.download('wordnet')
from nltk.stem.wordnet import WordNetLemmatizer


ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
# Getting a list of all token types
tokenTypes = tok.word_counts.keys()
# print(tokenTypes)

# Need to additionally check if there are existing lemmatized tokens in the list already. So using a list comprehension here is not accurate.
# stemmed = [ps.stem(token) for token in tokenTypes]
# print(stemmed)
# stemmedToks = tf.keras.preprocessing.text.Tokenizer(num_words=None, lower=True, split=" ")
# stemmedToks.fit_on_texts(stemmed)

# Lemmatizing
# lemms = [lemmatizer.lemmatize(token) for token in tokenTypes]

# print(f'Lemms: {lemms}')
# lemmedToks = tf.keras.preprocessing.text.Tokenizer(num_words=None, lower=True, split=" ")
# lemmedToks.fit_on_texts(lemms)

stemmed = []
lemms = []
for token in tokenTypes:
    stemmedTok = ps.stem(token)
    if stemmedTok not in stemmed:
        stemmed.append(stemmedTok)
    lemmedTok = lemmatizer.lemmatize(token)
    if lemmedTok not in lemms:
        lemms.append(lemmedTok)

print(f'Total number of types after stemming: {len(stemmed)}')
print(f'Total number of types after lemmatizing: {len(lemms)}')

# asserting
print(f'Types: {len(tok.word_counts)} > lemmatized: {len(lemms)} > stemmed: {len(stemmed)}')
assert len(tok.word_counts) > len(lemms) > len(stemmed)
```
