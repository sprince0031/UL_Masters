## Task 2.1

Downloading the shakespeare corpus for all our processing

```python
#TASK 2.1
!wget http://www.gutenberg.org/files/100/100-0.txt
filePath = "./100-0.txt"
shakespeareFile = open(filePath, "r")
shakespeareContent = shakespeareFile.read()
shakespeareContent = shakespeareContent.splitlines()
print(shakespeareContent[0:100])
```

## Task 2.2

Tokenising using Keras.

```python
import tensorflow as tf

# Initialising the tokeniser pre-processor
tok = tf.keras.preprocessing.text.Tokenizer(num_words=None, lower=True, split=" ")
tok.fit_on_texts(shakespeareContent)
print(tok.word_counts)
```

```python
# token count
print(f'a. Total token count in the corpus: {tok.document_count}')
print(f'b. Total token types in the corpus: {len(tok.word_counts)}')
typeListByMostFrequency = sorted(list(tok.word_counts.items()), key=lambda item: item[1], reverse=True)
print(f'c. Most frequent token type is "{typeListByMostFrequency[0][0]}" in the corpus which has a frequency of {typeListByMostFrequency[0][1]}.')
```

## Task 2.3

Stemming and lemmatizing

```python
from nltk.stem import PorterStemmer
import nltk
nltk.download('wordnet')
from nltk.stem.wordnet import WordNetLemmatizer


ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Getting a list of all token types
tokenTypes = [item[0] for item in typeListByMostFrequency]
# print(tokenTypes)

stemmed = [ps.stem(token) for token in tokenTypes]
# print(stemmed)
stemmedToks = tf.keras.preprocessing.text.Tokenizer(num_words=None, lower=True, split=" ")
stemmedToks.fit_on_texts(stemmed)
print(f'Total number of types after stemming: {len(stemmedToks.word_counts)}')

# Lemmatizing
lemms = [lemmatizer.lemmatize(token) for token in tokenTypes]
# print(f'Lemms: {lemms}')
lemmedToks = tf.keras.preprocessing.text.Tokenizer(num_words=None, lower=True, split=" ")
lemmedToks.fit_on_texts(lemms)
print(f'Total number of types after stemming: {len(lemmedToks.word_counts)}')

# asserting
print(f'Types: {len(tok.word_counts)} > stemmed: {len(stemmedToks.word_counts)} > lemmatized: {len(lemmedToks.word_counts)}')
```
